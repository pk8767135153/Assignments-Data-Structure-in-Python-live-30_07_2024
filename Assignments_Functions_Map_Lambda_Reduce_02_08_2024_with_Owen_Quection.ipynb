### 5 What are iterators in python and how do they differ from inerrable

```python
"""
    Iterables : 
            An iterable is any object that you can loop over or iterate through 
            Iterables have an __iter__() method that produces items on demand when you iterate over them.
            When you pass an iterable to the iter() function, it returns an iterator.
    
    Iterators : 
            An iterator is an object that keeps a state and produces the next value each time it's iterated upon.
            Iterators implments two methods:
                __iter__():Typically returns the iterator objects
                __next__(): Return the next item in the iteration.
    Iterables are objects you can loop over, while iterators control the iteration process.
    An iterable has an __iter__() method, where as an iterator has both __iter__() and __next__() methods.
    ## All iterators are iteralbles, but not all iterables are iterators. 
"""
```

```python
# Iterables
my_list = [1, 2, 3]
for item in my_list:
    print(item)
```

```python
# Iterators:
my_iter = iter(my_list)
print(next(my_iter))  # Output: 1
print(next(my_iter))  # Output: 2
```

### 6 Explain the concept of generators in python and how they are defined

```python
"""
    Generators : 
        Generator are special functions that behave like iterators
        Unlikes lists or tuples which store all values in memory, generators produce values on the fly as needed
        They are memory efficient and useful for large dataset or streaming data
        Generators having few advanced Generator methods that is .send(),.throw(),.close()
        In Generators fucntion we mainly faces StopIteration Error 
        
    Define Generators:
        Create a generator function using the yield keyword.
        When a generator function is called, it doesn't execute immediately.
        The genreator function can be paused and resumed during iteration.
"""
```

```python
letters = ["a", "b", "c", "y"]
it = iter(letters)
while True:
    try:
        letter = next(it)
    except StopIteration:
        break
    print(letter)
```

```python
def generator_function():
  yield 1
  yield 2
  yield 3
  yield 4
  yield 5

# Create the generator
gen = generator_function()
while True:
    print(next(gen))
```

### 5. Create an iterator class in Python that generates the Fibonacci Sequecnce up to a specified number of terms.

```python
class FibonacciIterator:
    def __init__(self, num_terms):
        self.num_terms = num_terms
        self.current_term = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.current_term >= self.num_terms:
            raise StopIteration
        result = self.a
        self.a, self.b = self.b, self.a + self.b
        self.current_term += 1
        return result

# Example usage:
num_terms = 8

fibonacci_gen = FibonacciIterator(num_terms)
for term in fibonacci_gen:
    print(term, end=' ')
```

### 6. Write a generator function in python that yields the powers of 2 up to a given exponent.

```python
def power_of_two(exponent):
    for p in range(1,exponent+1):
        yield 2**p

exponent = 4
for values in power_of_two(exponent):
    print(values, end = " ")
```
